```python
#Multiple Regression
#Importing the libraries
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('reklam.csv')
data.head()
```

```python
x = data.iloc[:,1:-1].values #bağımsız değişkenler (TV,radio,gazete)
```

```python
y =data.iloc[:,-1].values.reshape(-1,1) #bağımlı değişken (satışlar)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state = 22)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
yhead = lr.predict(xtest)   #yhead = ypredict tahmin edilen y değerleri
yhead
```

```python
lr.predict([[30,40,50]])
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(ytest,yhead))

r2= r2_score(ytest,yhead)

print('rmse değeri:',rmse)
print('r2 değeri:',r2)
```

```python
def calculate_rmse(predictions, targets):
    """
    Calculates the Root Mean Squared Error (RMSE) between the predicted values and the target values.

    Args:
        predictions (list): A list of predicted values.
        targets (list): A list of target values.

    Returns:
        float: The RMSE value.
    """
    # Calculate the squared differences between predictions and targets
    squared_diff = [(p - t) ** 2 for p, t in zip(predictions, targets)]

    # Calculate the mean squared difference
    mean_squared_diff = sum(squared_diff) / len(predictions)

    # Calculate the RMSE by taking the square root of the mean squared difference
    rmse = mean_squared_diff ** 0.5

    return rmse
```

```python
print(calculate_rmse(yhead,ytest))
```
