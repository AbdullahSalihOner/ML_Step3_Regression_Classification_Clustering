```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('maaş.csv')
```

```python
data.head()
```

```python
x = data.Tecrübe.values.reshape(-1,1)
y = data.Maaş.values.reshape(-1,1)
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
plt.scatter(x,y)
plt.plot(x,y,color='red')
plt.xlabel('tecrübe')
plt.ylabel('maaş')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # modeli oluşturduk
```

```python
lr.fit(xtrain,ytrain) # modeli eğittik
```

```python
yhead = lr.predict(xtest) # modeli test ettik   
```

```python
yhead
```

```python
plt.scatter(x,y,color='red')
plt.plot(xtest,yhead)
```

```python
plt.scatter(x,y,color='red')
plt.plot(xtest,yhead)
plt.scatter(xtest,ytest,color='blue')
```

```python
lr.predict([[11]])
```

```python
lr.predict([[0]])
```

```python
lr.coef_ # eğim
```

```python
lr.intercept_ # y eksenini kestiği nokta
```

```python
from sklearn.metrics import r2_score,mean_squared_error
#performans metrikleri, r2_score, mean_squared_error
rmse = np.sqrt(mean_squared_error(ytest,yhead))
r2 = r2_score(ytest,yhead)
print(f"RMSE: {rmse}, R2 Score: {r2}")
```

```python
#R^2 değeri 1'e ne kadar yakınsa model o kadar iyi demektir.
#RMSE değeri 0'a ne kadar yakınsa model o kadar iyi demektir.
#sonuç olarak modelimiz iyi bir modeldir.
```

```python
#Gerçek ve tahmin edilen veriyi çizdirme
plt.scatter(x, y, color = "blue", label = "Gerçek Veri")
plt.scatter(xtest, yhead, color = "red", label = "Tahmin Edilen Veriler")
plt.plot(xtest, yhead, color = "green", label = "Tahmin Çizgisi")
plt.xlabel("Tecrübe")
plt.ylabel("Maaş")
plt.legend()
plt.show()
```

```python
#Model katsayı değerleri
print("Katsayı (Eğim): ", lr.coef_)
print("Y-Intercept (Kesişim noktası): ", lr.intercept_)
```
